# Matplotlib and Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Scatter Plot

### Using Matplotlib

```python
np.random.seed(20)
plotData = {
    'Category_X': [1,2,3,4,5,6],
    'Value_Y': np.random.randint(1,100,6)
}
```

```python
dataframe = pd.DataFrame(plotData)
dataframe
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(dataframe['Category_X'], dataframe['Value_Y'], label='Data Points')
plt.title('Scatter Plot using Matplotlib')
plt.xlabel('Category_X')
plt.ylabel('Value_Y')
plt.legend()
plt.grid(True)
plt.show()
```

### Using Seaborn

```python
plt.figure(figsize=(6, 6))
sns.scatterplot(x='Category_X', y='Value_Y', data=dataframe, label='Data Points')
plt.title('Scatter Plot using Seaborn')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(False)
plt.show()
```

## Bar Plot

##### **np.random.randint(low=1, high=100, size=6)** will generate an array of 6 random integers between 1 and 100 (inclusive of 1 and exclusive of 100)

```python
np.random.seed(20)
np.random.randint(low=1, high=100, size=6)
```

```python
np.random.seed(20)
plotData = {
    'Category_X': ['A', 'B', 'C', 'D', 'E', 'F'],
    'Value_Y': np.random.randint(1,100,6)
}
```

```python
sns.barplot(x='Category_X', y='Value_Y', data=plotData)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```

---

##### The **enumerate()** function is used to **loop through the values list** and **get both the index and value**.

```python
np.random.seed(20)
sns.barplot(x='Category_X', y='Value_Y', data=plotData)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')

for index, value in enumerate(np.random.randint(1,100,6)):
    plt.text(index, value + 0.7, str(value), ha='center')
    # plt.text is used for text annotation at specific x and y coordinates (index, value + 0.7)
plt.show()
```

**index**: This parameter represents the x-coordinate where you want to place the text annotation. In a bar plot, each bar is assigned an index (0, 1, 2, ...) based on its position on the x-axis. This index value **specifies the position of the bar**.


**value + 0.7**: It determines the y-coordinate of the text annotation. By adding 0.7 to the value, it **adjusts the vertical position** of the annotation text slightly above the top of each bar.

**ha='center'**: This parameter specifies the horizontal alignment of the text. Setting it to 'center' aligns the text horizontally at the specified x-coordinate.

# Histogram Plot

The ***np.random.randn()*** function generates **random numbers from a standard normal distribution (mean 0, standard deviation 1)**. This means the values it **produces can range from negative infinity to positive infinity**, with a **mean centered around 0**.

The **np.random.rand()** function generates **random numbers from a uniform distribution between 0 and 1**. Therefore, the range of numbers returned by np.random.rand() is **from 0 (inclusive) to 1 (exclusive)**.

```python
np.random.seed(30)
np.random.randn(10)
```

```python
np.random.seed(40)
data = np.random.randn(1000)

sns.histplot(data)
plt.xlabel('Values')
plt.ylabel('Number of occurrences')
plt.title('Histogram Plot')
plt.show()
```

---

```python
np.random.seed(40)
data = np.random.randn(1000)

sns.kdeplot(data, fill = True, color = 'orange')
plt.xlabel('Values')
plt.ylabel('Density')
plt.title('Kernel Density Estimation (KDE) Plot')
plt.show()
```

## Histogram with Kernel Density Estimation (KDE)

```python
np.random.seed(40)
data = np.random.randn(1000)

sns.histplot(data, kde = True, color = 'green')
plt.xlabel('Values')
plt.ylabel('Number of occurrences')
plt.title('Kernel Density Estimation (KDE) Plot')
plt.show()
```

### Using Seaborn

```python
np.random.seed(40)
data = np.random.randn(1000)

#sns.boxplot(data,showfliers=False)
# showfliers = False excludes outliers while plotting

sns.boxplot(data)
plt.xlabel('Data')
plt.ylabel('Values')
plt.title('Boxplot using Seaborn')
plt.show()
```

---

```python
np.random.seed(40)
data = np.random.randn(1000)

# Creating boxplot
sns.boxplot(data, color='green')
plt.xlabel('Data')
plt.ylabel('Values')
plt.title('Boxplot using Seaborn')


# Calculating the limits for outliers
quartile1 = np.percentile(data, 25)
quartile2 = np.percentile(data, 50)
quartile3 = np.percentile(data, 75)

interQuartileRange = quartile3 - quartile1

# Upper limit for outliers
upperLimit = quartile3 + 1.5 * interQuartileRange

# Lower limit for outliers
lowerLimit = quartile1 - 1.5 * interQuartileRange

plt.text(0.5, quartile1, f'Q1: {quartile1:.3f}', ha='center', va='center', color='purple')
plt.text(0.5, quartile2, f'Q2: {quartile2:.3f}', ha='center', va='center', color='purple')
plt.text(0.5, quartile3, f'Q3: {quartile3:.3f}', ha='center', va='center', color='purple')


# Annotating the plot with outlier limits
plt.text(0.34, upperLimit, f'Upper Limit: {upperLimit:.3f}', ha='center', va='center', color='red')
plt.text(0.34, lowerLimit, f'Lower Limit: {lowerLimit:.3f}', ha='center', va='center', color='red')

plt.show()
```

## Subplots

**np.linspace()** is used to create an array of evenly spaced numbers over a specified range.

```python
x = np.linspace(0.1, 4 * np.pi,100)

y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = 1 / np.sin(x)
y5 = 1 / np.cos(x)
y6 = 1 / np.tan(x)
```

```python
figure, axes = plt.subplots(nrows = 6, ncols = 1, figsize=(9,15))

axes[0].set_title('Sine Function')
axes[0].plot(x,y1,color = 'green')

axes[1].set_title('Cos Function')
axes[1].plot(x,y2,color = 'red')

axes[2].set_title('Tan Function')
axes[2].plot(x,y3,color = 'blue')

axes[3].set_title('Cosine Function')
axes[3].plot(x,y4,color = 'purple')

axes[4].set_title('Secant Function')
axes[4].plot(x,y5,color = 'yellow')

axes[5].set_title('Cot Function')
axes[5].plot(x,y6,color = 'green')

plt.suptitle('Trignometric Function')
plt.xlabel('Angles')

plt.tight_layout()
plt.show()
```

## Pie Chart

```python
proportions = [20, 30, 25, 15, 10]
categories = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.figure(figsize=(8, 6))
plt.pie(proportions, labels= categories, autopct='%1.2f%%', startangle=90)
# autopct='%1.2f%%' adds percentage values to each wedge of the pie.
# startangle=90 rotates the starting angle of the pie plot to start from the vertical line (12 o'clock position).

plt.title('Pie Chart')
plt.show()
```
